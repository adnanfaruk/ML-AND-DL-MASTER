In this assignment students have to transform iris data into 3 dimensions
and plot a 3d chart with transformed dimensions and colour each data
point with specific class.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn.decomposition import PCA
from sklearn import datasets
import seaborn as sns
```

```python
data=sns.load_dataset('iris')
data.head()
```

```python
x=data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
x
```

```python
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(x)
principlDf=pd.DataFrame(data=principalComponents,columns=['principal component 1', 'principal component 2', 'principal component 3'])
```

```python
principlDf
```

```python

principlDf.insert(3,'Species',data['species'].values)
```

```python
principlDf
```

```python
principlDf['Species'].value_counts()
```

```python
fig=plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('principal component 1')
ax.set_ylabel('principal component 2')
ax.set_zlabel('principal component 3')
ax.set_title('$3 dimensions $')

Species=['setosa','versicolor','virginica']
colors=['r','g','b']

for Specie,color in zip(Species,colors):
    indicestokeep=principlDf['Species']==Specie
    ax.scatter(principlDf.loc[indicestokeep,'principal component 1'],
               principlDf.loc[indicestokeep,'principal component 2'],
               principlDf.loc[indicestokeep,'principal component 3'], c=color,s=50)
    ax.legend(Species)
    ax.grid()
    
    
```
